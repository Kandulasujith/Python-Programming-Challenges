```python
class Solution(object): #running sum of 1d array
    def runningSum(self, nums):
        lst=[]
        for x in range(len(nums)):
            lst.append(sum(nums[0:x+1]))
        return lst
s=Solution()
s.runningSum([1,2,3,4]) 
```

```python
class Solution(object): #shuffle the array
    def shuffle(self, nums, n):
        new=[]
        lst1,lst2=nums[:n],nums[n:]
        for x in range(len(lst1)):
            new.extend([lst1[x],lst2[x]])
        return new
ob=Solution()
ob.shuffle(nums = [1,1,2,2], n = 2)        
```

```python
class Solution(object): #kids with the greatest number of candies
    def kidsWithCandies(self, candies, extraCandies):
        highest=max(candies)
        lst=[]
        for x in candies:
            if (x+extraCandies)>=highest:
                lst.append('true')
            else:
                lst.append('false')
        return lst
obj=Solution()
print(obj.kidsWithCandies(candies = [12,1,12], extraCandies = 10))
```
